```python
import pandas as pd
import numpy as np
from IPython.display import display

# Simulating a broken dataset extract
data = {
    'student_id': [1, 2, 3, 4, 4, 5], # Notice the duplicate ID 4
    'name': ['Ali', 'Sara', 'Omar', 'Lina', 'Lina', np.nan], # Missing name at the end
    'score': [14.5, 9.0, np.nan, 18.5, 18.5, 15.0], # Missing score in the middle
    'subject': ['Math', 'Math', 'Physics', 'Physics', 'Physics', 'Math']
}

df_messy = pd.DataFrame(data)
print("--- Raw Messy Data ---")
display(df_messy)
```

```python
print('--- 1. Dropping Duplicates ---')

df_clean = df_messy.drop_duplicates().copy()
display(df_clean)

print("--- 2. Handling Missing Text (Names) ---")
df_clean["name"] = df_messy['name'].fillna("Unknown") # Replacing missing names with Unknown

print("--- 3. Handling Missing Numbers (Scores) ---")
average_score = df_clean['score'].mean()
df_clean['score'] = df_messy['score'].fillna(average_score) # Replace missing scores with average score

print("--- 4. The final result ---")
display(df_clean)
```

```python
print("--- Average score per subject ---")
subject_stats = df_clean.groupby('subject')['score'].mean().reset_index() # SQL equivalent: SELECT subject, AVG(score) FROM table GROUP BY subject
display(subject_stats)
```
