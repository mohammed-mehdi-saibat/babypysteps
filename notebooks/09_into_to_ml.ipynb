```python
import pandas as pd
from IPython.display import display

# 1. Our historical dataset
data = {
    'hours_studied': [1, 2, 3, 4, 5],
    'score': [20, 40, 60, 80, 100]
}
df = pd.DataFrame(data)

# 2. Splitting into X (Features) and y (Target)
X = df[["hours_studied"]] # Double brackets keeps it as a 2D DataFrame
y = df["score"] # Single bracket makes it a 1D Series

print("--- X (The Inputs) ---")
display(X)

print("\n--- y (The Answers) ---")
display(y)
```

```python
from sklearn.linear_model import LinearRegression

# 1. Initialize the algorithm (Just like instantiating an OOP class!)
model = LinearRegression()

# 2. Train the model (This is where the actual "learning" happens)
model.fit(X, y)

# 3. Make predictions on unseen data
# What if a student studies for 6 hours? Or 2.5 hours?
future_data = pd.DataFrame({'hours_studied': [6, 2.5]})
predictions = model.predict(future_data)

print(f"Predicted score for 6 hours: {predictions[0]}")
print(f"Predicted score for 2.5 hours: {predictions[1]}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 

# 1. Generating a dummy dataset of 100 students
# (We add some random noise so the data isn't a "perfect" line)
np.random.seed(42)
hours = np.random.uniform(1,10,100)
scores = (hours * 10) + np.random.normal(0,5,100)
df_large = pd.DataFrame({"hours_studied": hours, "score": scores})

X_large = df_large[["hours_studied"]]
y_large = df_large["score"]

# 2. The Train/Test Split!
# We tell Scikit-Learn to randomly slice off 20% of the data for testing
X_train, X_test, y_train, y_test = train_test_split(
    X_large,
    y_large,
    test_size = 0.2,
    random_state = 42 # This ensures the random split is the same everytime it's ran
)

print(f"Total records: {len(X_large)}")
print(f"Training records (80%): {len(X_train)}")
print(f"Testing records (20%): {len(X_test)}")
```

```python
from sklearn.metrics import mean_absolute_error

# 1. Train the model ONLY on the 80% training data
model_large = LinearRegression()
model_large.fit(X_train, y_train)

# 2. Give it the "Final Exam" - ask it to predict the 20% it has never seen
test_predictions = model_large.predict(X_test)

# 3. Grade the exam! 
# We compare what the model predicted vs the actual real answers (y_test)
error = mean_absolute_error(y_test, test_predictions)

print(f"On average, our model's predictions are off by: {error:.2f} points.")
```
