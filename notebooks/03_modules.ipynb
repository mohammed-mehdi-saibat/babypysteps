### The difference between executing and importing

The `if __name__ == "__main__":` statement acts like a shield:

1. **When we type `python main.py` in the terminal:** Python knows we are running the file directly. It lowers the shield and executes all the code inside.
2. **When we do an `import` in this Notebook:** Python keeps the shield up. It quietly imports our functions so we can use them, but **it refuses to automatically execute** the rest of the script.

This prevents us from accidentally running an entire program when we just want to borrow a single tool!
