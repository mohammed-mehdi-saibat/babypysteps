### Comparaison des Syntaxes : PHP vs Python

| Concept | PHP | Python |
| :--- | :--- | :--- |
| **Variables** | `$nom = "Ali";` | `nom = "Ali"` |
| **Tableaux simples** | `$arr = [1, 2, 3];` | `arr = [1, 2, 3]` |
| **Tableaux associatifs** | `['clé' => 'valeur']` | `{'clé': 'valeur'}` |
| **Concaténation** | `"Hello " . $name` | `f"Hello {name}"` |
| **Boucle d'itération** | `foreach ($items as $item)` | `for item in items:` |
| **Déclaration de fonction**| `function faire_quelque_chose()` | `def faire_quelque_chose():` |
| **Délimiteurs de blocs** | Accolades `{ }` | Indentation (4 espaces) |
| **Valeur nulle** | `null` | `None` |
| **Importation** | `use App\Models\User;` | `from module import Class` |
