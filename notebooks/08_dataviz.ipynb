```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Recreating our clean summary data from Day 3, plus a few extra subjects
data = {
    'subject': ['Math', 'Physics', 'History', 'Biology'],
    'average_score': [12.8, 16.4, 14.5, 15.2]
}
df_stats = pd.DataFrame(data)

print("--- Matplotlib (The foundation) ---")
plt.bar(df_stats['subject'], df_stats['average_score']) # Build a basic bar chart

# Add some labels
plt.title("Average Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Score")

plt.show() # Render the chart
```

```python
print("--- Seaborn (The professional look) ---")

sns.set_theme(style="whitegrid") # Apply  a modern theme to the background

sns.barplot(data=df_stats, x="subject", y="average_score", hue="subject", palette="viridis") # Create the bar chart with a built in color palette

plt.title("Average score by subject (Seaborn)")
plt.show()
```
